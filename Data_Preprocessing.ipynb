## Import and Preprocess Data

csv utf-8로 재저장해야 import 가능

```python
nextdate = '20230102'
data = pd.DataFrame()
for _ in range(27):
    nextdate = pd.to_datetime(nextdate, format='%Y%m%d') + timedelta(days=7)
    filename = str(nextdate)[0:10].replace('-','')+'.csv'
    path = "/Users/ce/Desktop/GSDS/2023_Summer/dacon_krx/rawdata/foreign/"+filename
    
    try:
        raw = pd.read_csv(path, encoding='utf-8')
    except:
        
        continue
#    raw = pd.read_csv(path, encoding='utf-8')
    data = pd.concat([data,raw])
```

```python
#exception
raw = pd.read_csv("/Users/ce/Desktop/GSDS/2023_Summer/rawdata/dacon_krx/foreign/20230530.csv", encoding='utf-8')
data = pd.concat([data,raw])
raw = pd.read_csv("/Users/ce/Desktop/GSDS/2023_Summer/rawdata/dacon_krx/foreign/20230502.csv", encoding='utf-8')
data = pd.concat([data,raw])
raw = pd.read_csv("/Users/ce/Desktop/GSDS/2023_Summer/rawdata/dacon_krx/foreign/20230125.csv", encoding='utf-8')
data = pd.concat([data,raw])
```
